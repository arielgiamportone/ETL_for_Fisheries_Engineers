 Antes de zarpar, asegurÃ©monos de tener todo listo. Importaremos algunas funciones Ãºtiles que hemos guardado en nuestra caja de herramientas (`src/data_utils.py`).

```python
import pandas as pd
import os
import sys
```

```python
# AÃ±adimos la carpeta 'src' al path para poder importar nuestros mÃ³dulos
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from src.data_utils import get_data, generate_sample_data, generate_ocean_data, generate_fleet_data
```

## 1.1 Pescando Datos de Desembarques

 Intentaremos conectarnos a una fuente de datos pÃºblica (simulada aquÃ­). Si la conexiÃ³n falla (Â¡a veces el mar estÃ¡ picado para las conexiones!), generaremos datos de ejemplo realistas para no detener nuestra exploraciÃ³n.

```python
# Definimos la ubicaciÃ³n de nuestros "caladeros" (URLs) y dÃ³nde guardaremos la "pesca" (archivos locales)
url_desembarques = "https://datos.agroindustria.gob.ar/dataset/pesca-maritima/desembarques" # URL ejemplo
filename_desembarques = "../data/raw/desembarques_pesqueros.csv"

# Usamos nuestra funciÃ³n 'get_data' para obtener los datos
df_desembarques = get_data(url_desembarques, filename_desembarques)

# Echemos un primer vistazo a nuestra captura
print("\n--- Muestra de Datos de Desembarques ---")
print(df_desembarques.head())
print(f"\nTotal de registros de desembarques: {len(df_desembarques)}")
```

## 1.2 Explorando las Condiciones del OcÃ©ano

 Ahora, obtendremos datos sobre el ambiente marino. Estos datos son cruciales porque factores como la temperatura del agua influyen directamente en dÃ³nde se encuentran los peces.

```python
filename_oceano = "../data/raw/datos_oceanograficos.csv"

# Generamos datos oceanogrÃ¡ficos simulados (en un caso real, podrÃ­amos usar una API de NOAA o Copernicus)
# Nota: La funciÃ³n get_data podrÃ­a adaptarse para manejar APIs si fuera necesario.
# Por ahora, generamos directamente.
if not os.path.exists(filename_oceano):
     print(f"Generando datos oceanogrÃ¡ficos de ejemplo en {filename_oceano}")
     df_oceano = generate_ocean_data(filename_oceano)
else:
     print(f"Cargando datos oceanogrÃ¡ficos desde {filename_oceano}")
     df_oceano = pd.read_csv(filename_oceano)


# Vistazo a las condiciones del mar
print("\n--- Muestra de Datos OceanogrÃ¡ficos ---")
print(df_oceano.head())
print(f"\nTotal de registros oceanogrÃ¡ficos: {len(df_oceano)}")
```

## 1.3 Inspeccionando la Flota Pesquera

 Finalmente, recolectaremos informaciÃ³n sobre los barcos. Â¿Son grandes o pequeÃ±os? Â¿Modernos o antiguos? Esto nos ayudarÃ¡ a entender la capacidad de pesca.

```python
filename_flota = "../data/raw/flota_pesquera.csv"

# Generamos datos simulados de la flota
if not os.path.exists(filename_flota):
    print(f"Generando datos de flota de ejemplo en {filename_flota}")
    df_flota = generate_fleet_data(filename_flota)
else:
    print(f"Cargando datos de flota desde {filename_flota}")
    df_flota = pd.read_csv(filename_flota)

# Vistazo a nuestra flota
print("\n--- Muestra de Datos de la Flota Pesquera ---")
print(df_flota.head())
print(f"\nTotal de registros de flota: {len(df_flota)}")
```
