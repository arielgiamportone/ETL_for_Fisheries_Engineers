```python
# Importaci贸n de bibliotecas necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import os
from datetime import datetime
import plotly.graph_objects as go
import plotly.express as px

# Configuraci贸n de visualizaci贸n
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Visualizaci贸n de la Estructura del Proyecto 
Primero, vamos a crear una visualizaci贸n de la estructura de nuestro proyecto:

```python
def crear_arbol_proyecto():
    """
    Crea una visualizaci贸n del 谩rbol de directorios del proyecto
    """
    estructura = {
        'ETL_for_Fisheries_Engineers': {
            'data': {
                'raw': ['desembarques_pesqueros.csv', 'datos_oceanograficos.csv', 'flota_pesquera.csv'],
                'processed': ['desembarques_transformados.csv', 'datos_enriquecidos.csv', 'pesquerias.db']
            },
            'notebooks': [
                '00_Configuracion_Entorno.ipynb',
                '01_Extraccion_Tesoros_Datos.ipynb',
                '02_Transformacion_Pulido_Gemas.ipynb',
                '03_Carga_Exhibicion_Hallazgos.ipynb',
                '04_Visualizacion_Exploratoria.ipynb'
            ],
            'src': ['__init__.py', 'data_utils.py', 'processing.py']
        }
    }
    
    G = nx.Graph()
    def agregar_nodos(dict_actual, padre=None):
        for k, v in dict_actual.items():
            G.add_node(k)
            if padre:
                G.add_edge(padre, k)
            if isinstance(v, dict):
                agregar_nodos(v, k)
            elif isinstance(v, list):
                for item in v:
                    G.add_node(item)
                    G.add_edge(k, item)
    
    agregar_nodos(estructura)
    
    plt.figure(figsize=(15, 10))
    pos = nx.spring_layout(G, k=1, iterations=50)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', 
            node_size=2000, font_size=8, font_weight='bold')
    plt.title('Estructura del Proyecto ETL para Datos Pesqueros')
    plt.show()

crear_arbol_proyecto()
```

## 2. Visualizaci贸n del Pipeline ETL 
Ahora, vamos a crear una visualizaci贸n interactiva del flujo de datos:

```python
def visualizar_pipeline_etl():
    """
    Crea una visualizaci贸n interactiva del pipeline ETL
    """
    # Definir las etapas del pipeline
    etapas = ['Extracci贸n', 'Transformaci贸n', 'Carga']
    
    # Crear el diagrama de flujo
    fig = go.Figure()
    
    # Agregar nodos para cada etapa
    for i, etapa in enumerate(etapas):
        fig.add_trace(go.Scatter(
            x=[i],
            y=[0],
            mode='markers+text',
            name=etapa,
            text=[etapa],
            marker=dict(size=40, symbol='circle'),
            hoverinfo='text',
            hovertext=f'Etapa: {etapa}'
        ))
    
    # Agregar flechas entre etapas
    for i in range(len(etapas)-1):
        fig.add_annotation(
            x=i,
            y=0,
            xref="x",
            yref="y",
            axref="x",
            ayref="y",
            ax=i+1,
            ay=0,
            showarrow=True,
            arrowhead=2
        )
    
    # Actualizar el dise帽o
    fig.update_layout(
        title="Pipeline ETL para Datos Pesqueros",
        showlegend=False,
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        plot_bgcolor='white'
    )
    
    fig.show()

visualizar_pipeline_etl()
```

## 3. Monitoreo del Proceso ETL 
Vamos a crear algunas m茅tricas y visualizaciones para monitorear nuestro proceso ETL:

```python
def monitorear_proceso_etl():
    """
    Crea visualizaciones para monitorear el proceso ETL
    """
    # Cargar datos procesados
    df_raw = pd.read_csv('data/raw/desembarques_pesqueros.csv')
    df_processed = pd.read_csv('data/processed/desembarques_transformados.csv')
    
    # Calcular m茅tricas
    metricas = {
        'Registros Originales': len(df_raw),
        'Registros Procesados': len(df_processed),
        'Columnas Originales': len(df_raw.columns),
        'Columnas Procesadas': len(df_processed.columns),
        'Valores Nulos Originales': df_raw.isnull().sum().sum(),
        'Valores Nulos Procesados': df_processed.isnull().sum().sum()
    }
    
    # Crear visualizaci贸n de m茅tricas
    fig = go.Figure(data=[
        go.Bar(name='Datos Originales', 
               x=['Registros', 'Columnas', 'Valores Nulos'],
               y=[metricas['Registros Originales'], 
                  metricas['Columnas Originales'],
                  metricas['Valores Nulos Originales']]),
        go.Bar(name='Datos Procesados', 
               x=['Registros', 'Columnas', 'Valores Nulos'],
               y=[metricas['Registros Procesados'], 
                  metricas['Columnas Procesadas'],
                  metricas['Valores Nulos Procesados']])
    ])
    
    fig.update_layout(
        title='M茅tricas del Proceso ETL',
        barmode='group',
        xaxis_title='M茅trica',
        yaxis_title='Cantidad'
    )
    
    fig.show()

monitorear_proceso_etl()
```

4. Visualizaci贸n de la Calidad de Datos 

```python
def visualizar_calidad_datos():
    """
    Crea visualizaciones para evaluar la calidad de los datos
    """
    # Cargar datos procesados
    df = pd.read_csv('data/processed/datos_enriquecidos.csv')
    
    # Calcular completitud por columna
    completitud = (1 - df.isnull().sum() / len(df)) * 100
    
    # Crear gr谩fico de completitud
    fig = px.bar(
        x=completitud.index,
        y=completitud.values,
        title='Completitud de Datos por Columna (%)',
        labels={'x': 'Columna', 'y': 'Completitud (%)'}
    )
    
    fig.show()

visualizar_calidad_datos()
```
